#data load

```python
path = 'data/cl/cl_m240_20130101_20131231.csv'
```

```python
data = pd.read_csv(path, sep = ';')
data['Date'] = data['CLOSE_DATE'].map(str) + ' ' + data['CLOSE_TIME']
data.index = pd.to_datetime(data['Date'])
data =  data[~data.index.duplicated(keep='first')]
data['dow']= pd.to_datetime(data['Date']).dt.dayofweek
data.head()
```

#plotly 4H

```python
fig = go.Figure(data=[go.Candlestick(x=data.Date,
                open=data['OPEN'],
                high=data['HIGH'],
                low=data['LOW'],
                close=data['CLOSE'])])
fig.update_layout(
    autosize=False,
    width= 900,
    height=800,)

fig.show()
```

#Bollinger bands

```python
# Define the parameters for the Bollinger Band calculation
ma_size = 50
bol_size = 2
# Calculate the SMA
data.insert(0, 'moving_average', data['CLOSE'].rolling(ma_size).mean())
# Calculate the upper and lower Bollinger Bands
data.insert(0, 'bol_upper', data['moving_average'] + data['CLOSE'].rolling(ma_size).std() * bol_size)
data.insert(0, 'bol_lower', data['moving_average'] - data['CLOSE'].rolling(ma_size).std() * bol_size)
# Remove the NaNs -> consequence of using a non-centered moving average
data.dropna(inplace=True)
data.head()
```

```python
data.index[-1]
```

```python
filter_day = 'D'
days_week = pd.date_range(data.index[0], data.index[-1], freq=filter_day).to_series().dt.dayofweek
days_week[days_week == 1 ].index

```

```python
for index, day in days_week.items():
   #print(index, day)
   if day == 1:
      print(index)
      # print(days_week.index)
```

```python
#plotly
fig = go.Figure(data=[go.Candlestick(x=data.Date,
                open=data['OPEN'],
                high=data['HIGH'],
                low=data['LOW'],
                close=data['CLOSE'])])

# Plot the three lines of the Bollinger Bands indicator
for parameter in ['moving_average', 'bol_lower', 'bol_upper']:
    fig.add_trace(go.Scatter(
        x = data.Date,
        y = data[parameter],
        showlegend = False,
        line_color = 'gray',
        mode='lines',
        line={'dash': 'dash'},
        marker_line_width=2, 
        marker_size=10,
        opacity = 0.8))
# Plot the monday vlines

for indexes, day in days_week.items():
    if day == 1:
        fig.add_vline(x=indexes, 
        line_width=1, 
        line_dash="dash", 
        line_color="green")


fig.update_layout(
    autosize=False,
    width= 900,
    height=800,)

fig.show()
```
